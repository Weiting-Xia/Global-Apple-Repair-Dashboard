```python
import pandas as pd
```

```python
category=pd.read_csv('category.csv')
category.rename(columns={'category_id':'Category_ID','category_name':'Category_Name'},inplace=True)
category
```

```python
products=pd.read_csv('products.csv').dropna()
products=pd.merge(products,category,on='Category_ID',how='left')
products
```

```python
sales=pd.read_csv('sales.csv').dropna()
sales.rename(columns={'product_id':'Product_ID','category_name':'Category_Name','sale_id':'Sale_ID','sale_date':'Sale_Date','store_id':'Store_ID','quantity':'Quantity'},inplace=True)
sales=pd.merge(sales,products,on='Product_ID',how='left')
sales
```

```python
stores=pd.read_csv('stores.csv')
sales=pd.merge(sales,stores,on='Store_ID',how='left')
sales
```

```python
warranty=pd.read_csv('warranty.csv')
warranty.rename(columns={'claim_id':'Claim_ID','sale_id':'Sale_ID','claim_date':'Claim_Date','store_id':'Store_ID','repair_status':'Repair_Status'},inplace=True)
sales_warranty=pd.merge(sales,warranty,on='Sale_ID',how='left')
sales_warranty
```

```python
sales_warranty.to_csv('Apple_Retail.csv',index=False)
```
